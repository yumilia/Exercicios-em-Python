# <font color='blue'> 6. WRITE A CODE THAT TAKES A TEMPERATURE INPUT FROM A USER IN CELSIUS AND CONVERTS IT INTO FAHRENHEIT </font> 

```python
celsius = float(input("Digite um valor para temperatura: "))
fahrenheit = 1.8*celsius + 32
print("O valor convertido de Celsius para Fahrenheit é", fahrenheit)
```

#  <font color='blue'>7. WRITE A CODE THAT TAKES THE LENGTH AND WIDTH OF A RECTANGE FROM THE USER AND PRINTS THE AREA AND PERIMETER  </font> 


```python
a = float(input("Digite a altura:"))
b = float(input("Digite a largura:"))
print("A área é:", a*b)
print("O perímetro é:", 2*(a+b))
```

#  <font color='blue'>8. WRITE A CODE THAT TAKES THE RADIUS OF A CIRCLE FROM THE USER AND PRINTS THE AREA, DIAMETER, AND CIRCUMFERENCE  </font> 


```python
import math
raio = int(input("Digite um valor para o raio:"))
print("A área é", math.pi*raio*raio)
print("O diámetro é", 2*raio)
print("A circunferência é", 2*math.pi*raio)
```
